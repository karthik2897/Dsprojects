```python
Match_title = []
Series = []
Place = []
Date = []
Time = []
```

```python
data= list(zip(Match_title,Series,Place,Date,Time))
```

```python
df_fixture= pd.DataFrame(data,columns=["Match_title","Series","Place","Date","Time"])
```

# 5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/

```python
drivers= webdriver.Chrome(r"C://chromedriver.exe")
drivers.get("https://github.com/")
```

```python
#Clicking on Explore
WebDriverWait(drivers, 5).until(EC.element_to_be_clickable((By.XPATH, '//div[@class="HeaderMenu HeaderMenu--logged-out position-fixed top-0 right-0 bottom-0 height-fit position-lg-relative d-lg-flex flex-justify-between flex-items-center flex-auto"]/nav/ul/li[4]/details/summary'))).click()
#Clicking on Trending
WebDriverWait(drivers, 5).until(EC.element_to_be_clickable((By.XPATH, '//div[@class="dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4"]/ul[2]/li[3]/a'))).click()
```

```python
Repository_title = []
Repository_desc = []
Contributors_count = []
Language_used = []
```

```python
urls=[i.get_attribute("href")for i in drivers.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")]
for url in urls:
    drivers.get(url)
#Repository_title    
    try:
        title=drivers.find_element_by_xpath('//strong[@class="mr-2 flex-self-stretch"]/a')
        Repository_title.append(title.text.strip(" ").replace('\n',''))
        
    except NoSuchElementException as e:
        Repository_title.append("-")
#Repository_desc
    try:
        desc=drivers.find_element_by_xpath('//p[@class="f4 mt-3"]')
        Repository_desc.append(desc.text.strip(" ").replace('\n',''))
        
    except NoSuchElementException as e:
        Repository_desc.append("-")  
        
#Contributors_count
    try:
        count=drivers.find_element_by_xpath('//span[@class="Counter "]')
        Contributors_count.append(count.text.strip(" ").replace('\n',''))
        
    except NoSuchElementException as e:
        Contributors_count.append("-") 

#Language_used
    try:
        lang=drivers.find_element_by_xpath('//span[@class="color-text-primary text-bold mr-1"]')
        Language_used.append(lang.text.strip(" ").replace('\n',''))
        
    except NoSuchElementException as e:
        Language_used.append("-")
```

```python
data= list(zip(Repository_title,Repository_desc,Contributors_count,Language_used))
```

```python
df_trending= pd.DataFrame(data,columns=["Repository_title","Repository_desc","Contributors_count","Language_used"])
```

# 6. Scrape the details of top 100 songs on billiboard.com.
Url = https://www.billboard.com/

```python

drivers= webdriver.Chrome(r"C://chromedriver.exe")
drivers.get("https://www.billboard.com/")
```

```python
WebDriverWait(drivers, 5).until(EC.element_to_be_clickable((By.XPATH, '//ul[@class="header__subnav__list display--flex flex--left-center flex-md--center-center"]/li[3]/a'))).click()
```

```python
Rank = []
Song_name = []
Artist_name = []
Last_week_rank = []
Peak_rank = []
Weeks_on_board = []
```

```python
rank = drivers.find_elements_by_xpath('//span[@class="chart-element__rank__number"]')
name = drivers.find_elements_by_xpath('//span[@class="chart-element__information__song text--truncate color--primary"]')
artist = drivers.find_elements_by_xpath('//span[@class="chart-element__information__artist text--truncate color--secondary"]')
last = drivers.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--last"]')
peak = drivers.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--peak"]')
weeks = drivers.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--week"]')
```

```python
#Appendinf Ranks
for i in rank:
        Rank.append(i.text)
#Appending Names        
for i in name:
    if(i.text != None):
        Song_name.append(i.text.strip(" ").replace('\n',''))   
    else:
        Song_name.append("-")
#Appending Artist        
for i in artist:
    Artist_name.append(i.text.strip(" ").replace('\n',''))
#Appending Last week Rank    
for i in last:
    if(i.text != None):
        Last_week_rank.append(i.text.strip(" ").replace('\n',''))   
    else:
        Last_week_rank.append("-")
#Appending Peak Rank    
for i in peak:
    if(i.text != None):
        Peak_rank.append(i.text.strip(" ").replace('\n',''))   
    else:
        Peak_rank.append("-")
#Appending No of Weeks         
for i in weeks:
    Weeks_on_board.append(i.text.strip(" ").replace('\n',''))    
    
```

```python
data= list(zip(Rank,Song_name,Artist_name,Last_week_rank,Peak_rank,Weeks_on_board))
```

```python
df_hot100= pd.DataFrame(data,columns=["Rank","Song_name","Artist_name","Last_week_rank","Peak_rank","Weeks_on_board"])
```

```python
df_hot100
```

# 9. Scrape the details most watched tv series of all time from imdb.com.
 Url = https://www.imdb.com/list/ls095964455/

```python
drivers= webdriver.Chrome(r"C://chromedriver.exe")
drivers.get("https://www.imdb.com/list/ls095964455/")
```

```python
url = "https://www.imdb.com/list/ls095964455/"
res=requests.get(url)
print(res)
```

```python
soup=BeautifulSoup(res.content,'html.parser')
```

```python
soup.title.text
```

```python
shows = soup.findAll('div',{'class':'lister-item mode-detail'})
```

```python
Name = []
Year_span = []
Genre = []
Run_time = []
Rating = []
Votes = []
for i in shows:
    Name.append(i.h3.a.text)
    Year_span.append(i.find("span",{"class":"lister-item-year text-muted unbold"}).text)
    Genre.append(i.find("span",{"class":"genre"}).text.strip(" ").replace('\n',''))
    Run_time.append(i.find("span",{"class":"runtime"}).text)
    Rating.append(i.find("span",{"class":"ipl-rating-star__rating"}).text)
    Votes.append(i.find("span",{"name":"nv"}).text)
```

```python
data= list(zip(Name,Year_span,Genre,Run_time,Rating,Votes))
```

```python
df_tv100= pd.DataFrame(data,columns=["Name","Year_span","Genre","Run_time","Rating","Votes"])
```
