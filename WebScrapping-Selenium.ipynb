Q6.Scrape data of first 100 sunglasses listings on flipkart.com. You have to
scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount %

```python
drivers.get("https://www.flipkart.com/")
```

```python
#Input Search
search_item=drivers.find_element_by_xpath('//input[@class="_3704LK"]')
search_item.send_keys("Sunglasses")

#Click Search Button
search_btn=drivers.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_btn.click()
```

```python
Company= []
Description = []
Price= []
Discount = []
```

```python
j=1
while j<=3:
        company_tags=drivers.find_elements_by_xpath('//div[@class="_2WkVRV"]')
        desc_tags=drivers.find_elements_by_xpath('//a[@class="IRpwTa"]')
        price_tags=drivers.find_elements_by_xpath('//div[@class="_30jeq3"]')
        discount_tags=drivers.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')

        for i in company_tags:
                Company.append(i.text)   
        for i in desc_tags:
            #if(i.text != None):
                Description.append(i.text.replace('(57)',''))   
            #else:
                #Description.append("Null")
        for i in price_tags:
             Price.append(i.text.replace('\n',''))
        for i in discount_tags:
             Discount.append(i.text)  
        try:
            drivers.find_elements_by_xpath('//div[@class="_2MImiq"]/nav/a')[-1].click()
        except:
            print("End")
        j+=1
```

```python
data= list(zip(Company,Description,Price,Discount))
```
